```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
b=np.array([10.0,17.0])
A=np.array([[6,-2],[11.5,-3.85]])
x=np.linalg.solve(A,b)
print(x)
```

```python
u=np.linspace(0,100,100)
y1=(-1/2)*(10.0-6*u)
y2=(-1/3.85)*(17.0-11.5*u)
plt.figure(figsize=[6,2])
plt.plot(u,y1)
plt.plot(u,y2)
plt.show()
```
