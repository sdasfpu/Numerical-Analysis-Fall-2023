```python
def powerMethod(A,x0,tol):
  x=x0
  lambd= np.matmul(x,np.matmul(A,x))/np.matmul(x,x)

  lambdOld=lambd+1
  count=0
  while np.abs(lambd-lambdOld) > tol:
    count+=1
    if count >= 100:
      print('Too many iterations!')
      return

    y=np.matmul(A,x)
    x=y/np.linalg.norm(y,np.inf)

    lambdOld=lambd
    lambd = np.matmul(x,np.matmul(A,x))/np.matmul(x,x)

  return x, lambd
```

```python
A=np.array([[2,1],[1,2]])
x=np.array([1,0])
powerMethod(A,x,1e-16)
```
